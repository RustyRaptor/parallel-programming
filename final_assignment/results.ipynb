```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./performance_metrics.csv')

df.head()
```

```python
for program in df['Program'].unique():
    program_df = df[df['Program'] == program]
    plt.plot(program_df['Array Size'], program_df['Execution Time'], label=program)

plt.xlabel('Array Size')
plt.ylabel('Execution Time')
plt.title('Execution Time vs. Array Size for Different Programs (all thread counts averaged)')
plt.legend()
plt.show()
```

```python
# Graph all except cuda for 4 threads
df_no_cuda = df[df['Program'] != 'cuda']
df_no_cuda_4t = df_no_cuda[df_no_cuda['Thread Count'] == 4]

for program in df_no_cuda_4t['Program'].unique():
    program_df = df_no_cuda_4t[df_no_cuda_4t['Program'] == program]
    plt.plot(program_df['Array Size'], program_df['Execution Time'], label=program)

plt.xlabel('Array Size')
plt.ylabel('Execution Time')
plt.title('Execution Time vs. Array Size for Different Programs (4 threads)')
plt.legend()
        
```

```python
# Graph all except cuda for 8 threads
df_no_cuda_8t = df_no_cuda[df_no_cuda['Thread Count'] == 8]

for program in df_no_cuda_8t['Program'].unique():
        program_df = df_no_cuda_8t[df_no_cuda_8t['Program'] == program]
        plt.plot(program_df['Array Size'], program_df['Execution Time'], label=program)

plt.xlabel('Array Size')
plt.ylabel('Execution Time')
plt.title('Execution Time vs. Array Size for Different Programs (8 threads)')
plt.legend()

plt.show()
```

```python
# Graph all except cuda for 16 threads
df_no_cuda_16t = df_no_cuda[df_no_cuda['Thread Count'] == 16]

for program in df_no_cuda_16t['Program'].unique():
        program_df = df_no_cuda_16t[df_no_cuda_16t['Program'] == program]
        plt.plot(program_df['Array Size'], program_df['Execution Time'], label=program)

plt.xlabel('Array Size')
plt.ylabel('Execution Time')
plt.title('Execution Time vs. Array Size for Different Programs (16 threads)')
plt.legend()

plt.show()

```

```python
# Graph all except cuda for 32 threads
df_no_cuda_16t = df_no_cuda[df_no_cuda['Thread Count'] == 32]

for program in df_no_cuda_16t['Program'].unique():
        program_df = df_no_cuda_16t[df_no_cuda_16t['Program'] == program]
        plt.plot(program_df['Array Size'], program_df['Execution Time'], label=program)

plt.xlabel('Array Size')
plt.ylabel('Execution Time')
plt.title('Execution Time vs. Array Size for Different Programs (32 threads)')
plt.legend()

plt.show()

```

```python
# compare cuda to OpenMP
df_cuda = df[df['Program'] == 'cuda']
df_openmp = df[df['Program'] == 'openmp']

for thread_count in df_cuda['Thread Count'].unique():
        cuda_df = df_cuda[df_cuda['Thread Count'] == thread_count]
        openmp_df = df_openmp[df_openmp['Thread Count'] == thread_count]
        plt.plot(cuda_df['Array Size'], cuda_df['Execution Time'], label='cuda')
        plt.plot(openmp_df['Array Size'], openmp_df['Execution Time'], label='openmp')
        
        plt.xlabel('Array Size')
        plt.ylabel('Execution Time')
        plt.title('Execution Time vs. Array Size for Different Programs ({} threads)'.format(thread_count))
        plt.legend()
        plt.show()
```
