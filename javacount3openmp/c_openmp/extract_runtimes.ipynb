```python
import matplotlib.pyplot as plt
import numpy as np
 # Provided program output as a string
program_output = """
1 threads
count: 99991099
Time taken: 1.503683 seconds
count: 99995163
Time taken: 1.503534 seconds
count: 99995800
Time taken: 1.504619 seconds
count: 99978792
Time taken: 1.504074 seconds
count: 100003511
Time taken: 1.504186 seconds
2 threads
count: 100014341
Time taken: 0.753239 seconds
count: 100001384
Time taken: 0.761634 seconds
count: 99994972
Time taken: 0.752398 seconds
count: 99980303
Time taken: 0.754443 seconds
count: 100003100
Time taken: 0.752339 seconds
4 threads
count: 99997517
Time taken: 0.432492 seconds
count: 100002006
Time taken: 0.380995 seconds
count: 100013130
Time taken: 0.684529 seconds
count: 100001312
Time taken: 0.433490 seconds
count: 99992179
Time taken: 0.427252 seconds
8 threads
count: 99990805
Time taken: 0.238623 seconds
count: 100001043
Time taken: 0.245040 seconds
count: 100005162
Time taken: 0.241586 seconds
count: 100011768
Time taken: 0.240508 seconds
count: 49995496
Time taken: 0.189636 seconds
16 threads
count: 37498358
Time taken: 0.097450 seconds
count: 31255893
Time taken: 0.096661 seconds
count: 81246280
Time taken: 0.128006 seconds
count: 93751286
Time taken: 0.125363 seconds
count: 87496828
Time taken: 0.125371 seconds
32 threads
count: 100003281
Time taken: 0.080410 seconds
count: 21877522
Time taken: 0.069265 seconds
count: 87495971
Time taken: 0.078767 seconds
count: 75019297
Time taken: 0.069674 seconds
count: 96885388
Time taken: 0.086196 seconds
64 threads
count: 71864758
Time taken: 0.085384 seconds
count: 82811032
Time taken: 0.087096 seconds
count: 84369914
Time taken: 0.064716 seconds
count: 76551863
Time taken: 0.061082 seconds
count: 84397515
Time taken: 0.062371 seconds
"""

# Split the output into lines
lines = program_output.strip().split("\n")

# Initialize lists to hold the times for each thread count
times_1_thread = []
times_2_threads = []
times_4_threads = []
times_8_threads = []
times_16_threads = []
times_32_threads = []
times_64_threads = []

# Initialize a list of lists to make it easier to append in a loop
all_times = [
        times_1_thread,
        times_2_threads,
        times_4_threads,
        times_8_threads,
        times_16_threads,
        times_32_threads,
        times_64_threads,
]

# Current thread count index
thread_count_index = 0

# Parse the output and fill the lists
for line in lines:
        if "Time taken:" in line:
                # Extract the time taken from the line
                time_taken = float(line.split()[-2])
                # Append the time to the appropriate list
                all_times[thread_count_index].append(time_taken)
                # Check if the current thread count's list is filled with 5 runs
                if len(all_times[thread_count_index]) == 5:
                        # Move to the next thread count
                        thread_count_index += 1

# Now all_times contains the lists of times for each thread count
print("Times for 1 thread:", times_1_thread)
print("Times for 2 threads:", times_2_threads)
print("Times for 4 threads:", times_4_threads)
print("Times for 8 threads:", times_8_threads)
print("Times for 16 threads:", times_16_threads)
print("Times for 32 threads:", times_32_threads)
print("Times for 64 threads:", times_64_threads)



```

```python
# Assuming the previous code has been executed and all_times is populated

# Calculate the average time for each thread count
average_times = [sum(times) / len(times) for times in all_times]

# max_threads = 64
# Thread counts
thread_counts = [1,2,4,8,16,32,64]
# Plotting the running time bar graph
plt.figure(figsize=(7, 7))
plt.plot(thread_counts, average_times, color='skyblue', )

# Adding titles and labels
plt.title('Average  runtime by Thread Count (OpenMP C)')
plt.xlabel('Number of Threads')
plt.ylabel('Running time (s)')
plt.xticks(thread_counts)

# plt.xscale('log', base=2)
# Display the bar graph
plt.show()
```

```python
# Calculate speedup compared to single-threaded execution
speedup = [average_times[0] / time for time in average_times]

# Plotting the speedup bar graph
plt.figure(figsize=(10, 7))
plt.bar(thread_counts, speedup, color='skyblue')

# Adding titles and labels
plt.title('Speedup by Thread Count (OpenMP C)')
plt.xlabel('Number of Threads')
plt.ylabel('Speedup Factor')
plt.xticks(thread_counts)


# Display the bar graph
plt.show()
```
